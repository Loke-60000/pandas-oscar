```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (15, 5)

pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60)
```

```python
"""
data format:

year,category,winner,entity
1927,ACTOR,False,Richard Barthelmess
1927,ACTOR,True,Emil Jannings
1927,ACTRESS,False,Louise Dresser

"""

df = pd.read_csv('raw_data/oscar_all_nominations.csv', on_bad_lines='skip')

number_of_categories = df['category'].nunique()
number_of_entities = df['entity'].nunique()
actress_categories_df = df[df['category'].str.contains("ACTRESS")]
actor_categories_df = df[df['category'].str.contains("ACTOR")]

# for females
number_of_unique_actress_entities = actress_categories_df['entity'].nunique()
number_of_actress_entities = actress_categories_df['entity'].count()

# for males
number_of_unique_actor_entities = actor_categories_df['entity'].nunique()
number_of_actor_entities = actor_categories_df['entity'].count()

number_of_unique_categories_per_year = df.groupby(['year','category']).nunique()

```

```python
print("Number of unique categories: ", number_of_categories)
print("Number of entities: ", number_of_entities)
print("Number of unique actress: ", number_of_unique_actress_entities)
print("Number of actress: ", number_of_actress_entities)
print("Number of unique actors: ", number_of_unique_actor_entities)
print("Number of actors: ", number_of_actor_entities)
```

```python

print("Number of unique categories per year: ", number_of_unique_categories_per_year)

ax = number_of_unique_categories_per_year.plot(kind='bar', figsize=(20, 20), legend=False, width=0.8)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Unique Categories')
ax.set_title('Number of Unique Categories per Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
